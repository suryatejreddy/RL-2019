```python
from ipynb.fs.full.Base import MultiArmBandit
import matplotlib.pyplot as plt
```

# Stationary Setting

```python
ucb = MultiArmBandit(upper_bound = True, steps = 1000, alpha = 0.1)
ucb.play(2000)

realistic_epsilon = MultiArmBandit(epsilon = 0.1, init_q = 0, steps = 1000, alpha = 0.1)
realistic_epsilon.play(2000)

optimistic_greedy = MultiArmBandit(epsilon = 0, init_q = 5, steps = 1000, alpha = 0.1)
optimistic_greedy.play(2000)
```

```python
plt.plot(realistic_epsilon.overall_average_rewards, label = 'Realistic, E Greedy (Q = 0, E = 0.1)')
plt.plot(optimistic_greedy.overall_average_rewards, label = 'Optimistic, Greedy, (Q = 5, E = 0)')
plt.plot(ucb.overall_average_rewards, label = 'UCB (C = 2)')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('Average Rewards')
plt.legend()
```

# Non Stationary

```python
ucb_ns = MultiArmBandit(upper_bound = True, stationary = False, steps = 1000, alpha = 0.1)
ucb_ns.play(2000)

realistic_epsilon_ns = MultiArmBandit(epsilon = 0.1, init_q = 0, stationary = False, steps = 1000, alpha = 0.1)
realistic_epsilon_ns.play(2000)

optimistic_greedy_ns = MultiArmBandit(epsilon = 0, init_q = 5, stationary = False, steps = 1000, alpha = 0.1)
optimistic_greedy_ns.play(2000)
```

```python
plt.plot(realistic_epsilon_ns.overall_average_rewards, label = 'Realistic, E Greedy (Q = 0, E = 0.1)')
plt.plot(optimistic_greedy_ns.overall_average_rewards, label = 'Optimistic, Greedy, (Q = 5, E = 0)')
plt.plot(ucb_ns.overall_average_rewards, label = 'UCB (C = 2)')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('% Average Rewards')
plt.legend()
```

# Analysis

### As we can see from the graph UCB performs better than greedy in a stationary setting but is optmistic still outperforms the two. UCB favors actions which aren't picked and hence the spikes in the plot. Optimistic seems to be a better solution for stationary.

### UCB is the worst amongs the three techniques as shown in the graph when it comes to non stationary. Because in a non stationary setting, exploration is favored and UCB limits its exploration with the passage of time and hence cannot get good estimates of the dynamic model.
