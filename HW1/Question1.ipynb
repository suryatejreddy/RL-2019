```python
from ipynb.fs.full.Base import MultiArmBandit
import matplotlib.pyplot as plt
```

# Stationary Setting

```python
greedy = MultiArmBandit(epsilon = 0)
greedy.play()

epsilon_greedy_small = MultiArmBandit(epsilon = 0.01)
epsilon_greedy_small.play()

epsilon_greedy_large = MultiArmBandit(epsilon = 0.1)
epsilon_greedy_large.play()
```

```python
plt.plot(greedy.overall_average_rewards, label = 'Greedy')
plt.plot(epsilon_greedy_small.overall_average_rewards, label = 'Epsilon Greedy E = 0.01')
plt.plot(epsilon_greedy_large.overall_average_rewards, label = 'Epsilon Greedy E = 0.1')
plt.xlabel('Steps')
plt.ylabel('Average Reward')
plt.legend()
```

```python
plt.plot(greedy.overall_percentage_optimal, label = 'Greedy')
plt.plot(epsilon_greedy_small.overall_percentage_optimal, label = 'Epsilon Greedy E = 0.01')
plt.plot(epsilon_greedy_large.overall_percentage_optimal, label = 'Epsilon Greedy E = 0.1')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('% Optimal Action')
plt.legend()
```
