```python
from ipynb.fs.full.Base import MultiArmBandit
import matplotlib.pyplot as plt
```

# Figure 2.3

```python
realistic_epsilon = MultiArmBandit(epsilon = 0.1, init_q = 0, alpha = 0.1, steps = 2000)
realistic_epsilon.play(2000)

optimistic_greedy = MultiArmBandit(epsilon = 0, init_q = 5, alpha = 0.1, steps = 2000)
optimistic_greedy.play(2000)
```

```python
plt.plot(realistic_epsilon.overall_percentage_optimal, label = 'Realistic, E Greedy (Q = 0, E = 0.1)')
plt.plot(optimistic_greedy.overall_percentage_optimal, label = 'Optimistic, Greedy, (Q = 5, E = 0)')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('% Optimal Action')
plt.legend()
```

# Exercise 2.6

### In the beginning, each action is explored atleast once by all agents because each action picked first will lead to a low reward as compared to initial optimistic value. Now, there is a high probability of picking the optimal action (based on whichever gave the least dissapointing result) because you have picked every action atleast once.  This probability according to the graph is 40% as compared to a random 10%. All the agents would at this stage would pick the most optimal action, hence the spike in the graph. After picking the optimal at this instant (towards the start), the reward is lower and results in dissapointment again, so the agent moves on to exploring again, hence the dip. 

### This pattern keeps repeating (spike and dip) untill the initial bias is minimized and is almost negligible.

# Figure 2.3 for Non Stationary

```python
realistic_epsilon_ns = MultiArmBandit(epsilon = 0.1, init_q = 0 ,stationary = False, alpha = 0.1, steps = 2000)
realistic_epsilon_ns.play(2000)

optimistic_greedy_ns = MultiArmBandit(epsilon = 0, init_q = 5, stationary = False, alpha = 0.1, steps = 2000)
optimistic_greedy_ns.play(2000)
```

```python
plt.plot(realistic_epsilon_ns.overall_percentage_optimal, label = 'Realistic, E Greedy (Q = 0, E = 0.1)')
plt.plot(optimistic_greedy_ns.overall_percentage_optimal, label = 'Optimistic, Greedy, (Q = 5, E = 0)')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('% Optimal Action')
plt.legend()
```

# Explanation for Stationary and Non Stationary

### In a stationary setting, optimistic method doesn't perform well initially because it is exploring, but overtime it performs better as exploitation is favored over exploration (by then good estimates would have been formed).

### But for non stationary settings, optimistic method doesn't perform well because it stops exploring after a file (picking actions which have good estimates (according to times picked)). Because the reward model keeps changing it ends up missing out on optimal actions eventually. This can be observed in the graph as well, that as the number of steps increase, the optimisic model deterioriates.
