```python
from ipynb.fs.full.Base import MultiArmBandit
import matplotlib.pyplot as plt
```

# Figure 2.3

```python
realistic_epsilon = MultiArmBandit(epsilon = 0.1, init_q = 0, alpha = 0.1)
realistic_epsilon.play()

optimistic_greedy = MultiArmBandit(epsilon = 0, init_q = 5, alpha = 0.1)
optimistic_greedy.play()
```

```python
plt.plot(realistic_epsilon.overall_percentage_optimal, label = 'Realistic, E Greedy (Q = 0, E = 0.1)')
plt.plot(optimistic_greedy.overall_percentage_optimal, label = 'Optimistic, Greedy, (Q = 5, E = 0)')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('% Optimal Action')
plt.legend()
```
