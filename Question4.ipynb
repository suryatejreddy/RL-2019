```python
from ipynb.fs.full.Base import MultiArmBandit
import matplotlib.pyplot as plt
```

# Stationary Setting

```python
ucb = MultiArmBandit(upper_bound = True)
ucb.play()

realistic_epsilon = MultiArmBandit(epsilon = 0.1, init_q = 0)
realistic_epsilon.play()

optimistic_greedy = MultiArmBandit(epsilon = 0, init_q = 5)
optimistic_greedy.play()
```

```python
plt.plot(realistic_epsilon.overall_average_rewards, label = 'Realistic, E Greedy (Q = 0, E = 0.1)')
plt.plot(optimistic_greedy.overall_average_rewards, label = 'Optimistic, Greedy, (Q = 5, E = 0)')
plt.plot(ucb.overall_average_rewards, label = 'UCB (C = 2)')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('% Average Rewards')
plt.legend()
```

# Non Stationary

```python
ucb_ns = MultiArmBandit(upper_bound = True, stationary = False)
ucb_ns.play()

realistic_epsilon_ns = MultiArmBandit(epsilon = 0.1, init_q = 0, stationary = False)
realistic_epsilon_ns.play()

optimistic_greedy_ns = MultiArmBandit(epsilon = 0, init_q = 5, stationary = False)
optimistic_greedy_ns.play()
```

```python
plt.plot(realistic_epsilon_ns.overall_average_rewards, label = 'Realistic, E Greedy (Q = 0, E = 0.1)')
plt.plot(optimistic_greedy_ns.overall_average_rewards, label = 'Optimistic, Greedy, (Q = 5, E = 0)')
plt.plot(ucb_ns.overall_average_rewards, label = 'UCB (C = 2)')
plt.tight_layout()
plt.xlabel('Steps')
plt.ylabel('% Average Rewards')
plt.legend()
```
